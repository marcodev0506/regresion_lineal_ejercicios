# Regresión lineal multivariable

## Importando datos

```python
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',
            header=None, sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head()
```

## Creando modelo con sklearn

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

X = df[['RM', 'INDUS']].values
y = df['MEDV'].values.reshape(-1, 1)
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)

slr = LinearRegression()
slr.fit(X_std, y_std)
```

## Análisis de resultados

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import *
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D

x1_range = np.arange(df['RM'].min(),df['RM'].max())
x2_range = np.arange(df['INDUS'].min(),df['INDUS'].max())

X1, X2 = np.meshgrid(x1_range,x2_range)

plano = pd.DataFrame({'RM':X1.ravel(), 'INDUS':X2.ravel()})
pred = slr.predict(plano).reshape(X1.shape)
pred = sc_y.inverse_transform(pred)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(X1, X2, pred, alpha=0.4)

# Graficar los puntos de dispersión
ax.scatter3D(df['RM'], df['INDUS'], df['MEDV'], color='r', marker='.')

# Ajustar la vista (opcional)
ax.view_init(elev=10, azim=5)

plt.show()
```

## Generando predicción con el modelo

```python
# Crea un array con el número de habitaciones y una característica ficticia
data = np.array([[5.0, 1]])  # 5 habitaciones y valor ficticio 1

# Ajusta el StandardScaler a ambas características
sc = StandardScaler()
sc.fit(data)

# Transforma los datos
data_std = sc.transform(data)

# Usa data_std para predecir el precio
price_std = slr.predict(data_std)
print('El precio en miles es de %.3f' % sc_y.inverse_transform(price_std))
```
